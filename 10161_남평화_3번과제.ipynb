```python

from datetime import datetime, date, timedelta


# 레거시 함수들 정의

def get_buildings():
    return building_ids

def get_floors(building_id):
    return {floor: floor_ids[building_id][floor][0] for floor in floor_ids[building_id]}

def get_rooms(building_id, floor_id):
    floors = get_floors(building_id)
    for floor_key in floors:
        if floors[floor_key] == floor_id:
            return floor_ids[building_id][floor_key][1]

def generate_reservation_id(building_id, floor_id, room_id, start_datetime):
    timestamp = start_datetime.strftime("%Y%m%d_%H%M")
    return f"{building_id}_{floor_id}_{room_id}_{timestamp}"

def parse_reservation_id(reservation_id):
    parts = reservation_id.split('_')
    if len(parts) >= 4:
        building_id = int(parts[0])
        floor_id = int(parts[1])
        room_id = int(parts[2])
        return building_id, floor_id, room_id
    return None, None, None

def check_time_overlap(building_id, floor_id, room_id, new_start, new_end, exclude_reservation_id=None):
    room_reservations = reservation_dict[building_id][floor_id][room_id]

    for res_id, reservation in room_reservations.items():
        if exclude_reservation_id and res_id == exclude_reservation_id:
            continue

        existing_start = reservation['start_datetime']
        existing_end = reservation['end_datetime']

        if new_start < existing_end and new_end > existing_start:
            return True, res_id

    return False, None

def add_reservation(building_id, floor_id, room_id, user_name, purpose, title, start_datetime, end_datetime):
    reservation_id = generate_reservation_id(building_id, floor_id, room_id, start_datetime)

    is_overlap, conflicting_reservation_id = check_time_overlap(
        building_id, floor_id, room_id, start_datetime, end_datetime
    )

    if is_overlap:
        return False, f"예약이 겹칩니다. 충돌하는 예약: {conflicting_reservation_id}", None
    else:
        reservation_dict[building_id][floor_id][room_id][reservation_id] = {
            'building_id': building_id,
            'floor_id': floor_id,
            'room_id': room_id,
            'user_name': user_name,
            'purpose': purpose,
            'title': title,
            'start_datetime': start_datetime,
            'end_datetime': end_datetime
        }
        return True, "예약이 성공적으로 추가되었습니다.", reservation_id

def cancel_reservation(reservation_id):
    building_id, floor_id, room_id = parse_reservation_id(reservation_id)

    if building_id is None:
        return False, "잘못된 예약 ID 형식입니다."

    if (building_id in reservation_dict and
        floor_id in reservation_dict[building_id] and
        room_id in reservation_dict[building_id][floor_id] and
        reservation_id in reservation_dict[building_id][floor_id][room_id]):

        del reservation_dict[building_id][floor_id][room_id][reservation_id]
        return True, "예약이 성공적으로 취소되었습니다."

    return False, "존재하지 않는 예약입니다."

def get_room_reservations(building_id, floor_id, room_id, date=None):
    if building_id not in reservation_dict:
        return []
    if floor_id not in reservation_dict[building_id]:
        return []
    if room_id not in reservation_dict[building_id][floor_id]:
        return []

    room_reservations = []

    for res_id, reservation in reservation_dict[building_id][floor_id][room_id].items():
        if date:
            if reservation['start_datetime'].date() != date:
                continue

        room_reservations.append({
            'reservation_id': res_id,
            **reservation
        })

    room_reservations.sort(key=lambda x: x['start_datetime'])
    return room_reservations

def update_reservation(reservation_id, **kwargs):
    building_id, floor_id, room_id = parse_reservation_id(reservation_id)

    if building_id is None:
        return False, "잘못된 예약 ID 형식입니다."

    if (building_id not in reservation_dict or
        floor_id not in reservation_dict[building_id] or
        room_id not in reservation_dict[building_id][floor_id] or
        reservation_id not in reservation_dict[building_id][floor_id][room_id]):
        return False, "존재하지 않는 예약입니다."

    old_reservation = reservation_dict[building_id][floor_id][room_id][reservation_id].copy()
    new_reservation = old_reservation.copy()
    new_reservation.update(kwargs)

    time_changed = ('start_datetime' in kwargs or 'end_datetime' in kwargs)
    room_changed = ('building_id' in kwargs or 'floor_id' in kwargs or 'room_id' in kwargs)

    if time_changed or room_changed:
        new_reservation_id = None
        if time_changed or room_changed:
            new_reservation_id = generate_reservation_id(
                new_reservation['building_id'],
                new_reservation['floor_id'],
                new_reservation['room_id'],
                new_reservation['start_datetime']
            )

        is_overlap, conflicting_reservation_id = check_time_overlap(
            new_reservation['building_id'],
            new_reservation['floor_id'],
            new_reservation['room_id'],
            new_reservation['start_datetime'],
            new_reservation['end_datetime'],
            exclude_reservation_id=reservation_id
        )

        if is_overlap:
            return False, f"예약이 겹칩니다. 충돌하는 예약: {conflicting_reservation_id}"

        del reservation_dict[building_id][floor_id][room_id][reservation_id]

        final_reservation_id = new_reservation_id if new_reservation_id else reservation_id
        reservation_dict[new_reservation['building_id']][new_reservation['floor_id']][new_reservation['room_id']][final_reservation_id] = new_reservation

        return True, f"예약이 성공적으로 수정되었습니다. 새 예약 ID: {final_reservation_id}"
    else:
        reservation_dict[building_id][floor_id][room_id][reservation_id] = new_reservation
        return True, "예약이 성공적으로 수정되었습니다."

def get_reservation(reservation_id):
    building_id, floor_id, room_id = parse_reservation_id(reservation_id)

    if building_id is None:
        return None

    if (building_id in reservation_dict and
        floor_id in reservation_dict[building_id] and
        room_id in reservation_dict[building_id][floor_id] and
        reservation_id in reservation_dict[building_id][floor_id][room_id]):

        reservation = reservation_dict[building_id][floor_id][room_id][reservation_id]
        return {'reservation_id': reservation_id, **reservation}

    return None

```

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage, ToolMessage
from langchain_core.tools import tool
from datetime import datetime, timedelta
import re

api_key = "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"

# LLM 초기화
llm = ChatOpenAI(
    model="gpt-4o",
    temperature=0.1,
    api_key=api_key,
)

# 사용자 이름
current_user = "사용자"

# Helper 함수: 건물 이름을 building_id로 변환
def get_building_id(building_name: str) -> int:
    """건물 이름을 building_id로 변환
    "sku타워", "sku-tower" 등 다양한 형식을 처리합니다.
    """
    building_name_lower = building_name.lower().replace(" ", "-").replace("타워", "-tower")

    # 직접 매칭
    if building_name_lower in building_ids:
        return building_ids[building_name_lower]

    # 부분 매칭 시도
    for key in building_ids:
        # "sku" -> "sku-tower" 매칭
        if building_name_lower in key or key.replace("-tower", "") in building_name_lower:
            return building_ids[key]

    # 한글 매칭 시도
    building_mapping = {
        "sku": "sku-tower",
        "skt": "skt-tower",
        "project107": "project-107",
        "skaxa": "skax-A",
        "skaxb": "skax-B",
    }

    for k, v in building_mapping.items():
        if k in building_name_lower:
            return building_ids[v]

    return None

# Helper 함수: 실제 층 번호를 내부 층 번호로 변환
def get_internal_floor_id(building_id: int, actual_floor_number: int) -> int:
    """실제 층 번호를 내부 층 번호로 변환합니다.
    floor_ids에서 [] 밖의 숫자가 실제 층이고, [] 안의 첫 번째 숫자가 내부 층 번호입니다.
    예: floor_ids[15473][24] = [28, {...}] 에서 실제 층은 24, 내부 층 번호는 28
    """
    if building_id not in floor_ids:
        return None

    # floor_ids에서 실제 층 번호 찾기
    if actual_floor_number in floor_ids[building_id]:
        return floor_ids[building_id][actual_floor_number][0]  # 내부 층 번호 반환
    return None

# Helper 함수: 사용자의 예약 찾기
def find_user_reservations(user_name: str, time_hint: str = None) -> list:
    """사용자의 예약을 찾습니다. time_hint가 있으면 시간으로 필터링합니다."""
    user_reservations = []

    for building_id in reservation_dict:
        for floor_id in reservation_dict[building_id]:
            for room_id in reservation_dict[building_id][floor_id]:
                for res_id, reservation in reservation_dict[building_id][floor_id][room_id].items():
                    if reservation['user_name'] == user_name:
                        if time_hint:
                            # 시간 힌트 파싱 (예: "6시", "7시", "오후 7시", "오전 7시")
                            hour_match = re.search(r'(\d+)시', time_hint)
                            if hour_match:
                                target_hour_12 = int(hour_match.group(1))

                                # 오후/오전 키워드 확인
                                time_hint_lower = time_hint.lower()
                                is_pm = any(keyword in time_hint_lower for keyword in ['오후', 'pm', '저녁', '밤', '야간'])
                                is_am = any(keyword in time_hint_lower for keyword in ['오전', 'am', '아침', '새벽'])

                                # 가능한 시간 목록 생성
                                possible_hours = []
                                if is_pm:
                                    # 오후면 12시간 형식으로 변환 (1시~11시는 +12, 12시는 그대로)
                                    if target_hour_12 == 12:
                                        possible_hours.append(12)
                                    else:
                                        possible_hours.append(target_hour_12 + 12)
                                elif is_am:
                                    # 오전이면 12시간 형식 (12시는 0시로)
                                    if target_hour_12 == 12:
                                        possible_hours.append(0)
                                    else:
                                        possible_hours.append(target_hour_12)
                                else:
                                    # 키워드가 없으면 오전/오후 모두 확인
                                    # 오전 시간
                                    if target_hour_12 == 12:
                                        possible_hours.append(0)
                                    else:
                                        possible_hours.append(target_hour_12)
                                    # 오후 시간 (1~11시만)
                                    if 1 <= target_hour_12 <= 11:
                                        possible_hours.append(target_hour_12 + 12)

                                # 예약 시간이 가능한 시간 목록에 있는지 확인
                                if reservation['start_datetime'].hour in possible_hours:
                                    user_reservations.append({
                                        'reservation_id': res_id,
                                        **reservation
                                    })
                        else:
                            user_reservations.append({
                                'reservation_id': res_id,
                                **reservation
                            })

    return user_reservations

# Tool 정의
@tool
def list_rooms(building_name: str, floor_number: int) -> str:
    """특정 건물의 특정 층에 있는 회의실 목록을 조회합니다.

    Args:
        building_name: 건물 이름 (예: "sku-tower", "skt-tower")
        floor_number: 층 번호 (예: 24)

    Returns:
        해당 층의 회의실 목록을 문자열로 반환
    """
    building_id = get_building_id(building_name)
    if building_id is None:
        return f"건물 '{building_name}'을 찾을 수 없습니다."

    # 실제 층 번호를 내부 층 번호로 변환
    internal_floor_id = get_internal_floor_id(building_id, floor_number)
    if internal_floor_id is None:
        return f"{building_name}의 {floor_number}층을 찾을 수 없습니다."

    rooms = get_rooms(building_id, internal_floor_id)
    if rooms is None:
        return f"{building_name} {floor_number}층에 회의실이 없습니다."

    room_list = ", ".join(rooms.keys())
    return f"{building_name} {floor_number}층 회의실 목록: {room_list}"

@tool
def create_reservation(
    building_name: str,
    floor_number: int,
    room_name: str,
    user_name: str,
    title: str,
    purpose: str,
    start_datetime: str,
    end_datetime: str
) -> str:
    """회의실 예약을 생성합니다.

    Args:
        building_name: 건물 이름
        floor_number: 층 번호
        room_name: 회의실 이름 (예: "2402-A")
        user_name: 예약자 이름
        title: 회의 제목
        purpose: 회의 목적
        start_datetime: 시작 시간 (형식: "YYYY-MM-DD HH:MM")
        end_datetime: 종료 시간 (형식: "YYYY-MM-DD HH:MM")

    Returns:
        예약 결과 메시지
    """
    building_id = get_building_id(building_name)
    if building_id is None:
        return f"건물 '{building_name}'을 찾을 수 없습니다."

    # 실제 층 번호를 내부 층 번호로 변환
    internal_floor_id = get_internal_floor_id(building_id, floor_number)
    if internal_floor_id is None:
        return f"{building_name}의 {floor_number}층을 찾을 수 없습니다."

    rooms = get_rooms(building_id, internal_floor_id)
    if rooms is None or room_name not in rooms:
        return f"회의실 '{room_name}'을 찾을 수 없습니다."

    room_id = rooms[room_name]

    try:
        start_dt = datetime.strptime(start_datetime, "%Y-%m-%d %H:%M")
        end_dt = datetime.strptime(end_datetime, "%Y-%m-%d %H:%M")
    except ValueError:
        return "날짜 형식이 올바르지 않습니다. 'YYYY-MM-DD HH:MM' 형식을 사용해주세요."

    success, message, reservation_id = add_reservation(
        building_id, internal_floor_id, room_id, user_name, purpose, title, start_dt, end_dt
    )

    return message

@tool
def update_user_reservation(
    time_hint: str,
    user_name: str,
    start_datetime: str = None,
    end_datetime: str = None,
    title: str = None,
    purpose: str = None
) -> str:
    """사용자의 예약을 수정합니다. 시간 힌트로 예약을 찾습니다.

    Args:
        time_hint: 시간 힌트 (예: "6시", "7시") - 수정할 예약을 찾기 위한 힌트
        user_name: 예약자 이름
        start_datetime: 새로운 시작 시간 (형식: "YYYY-MM-DD HH:MM", 예: "2025-12-04 19:00")
        end_datetime: 새로운 종료 시간 (형식: "YYYY-MM-DD HH:MM", 예: "2025-12-04 20:00")
        title: 새로운 회의 제목 (선택)
        purpose: 새로운 회의 목적 (선택)

    Returns:
        수정 결과 메시지
    """
    reservations = find_user_reservations(user_name, time_hint)

    if len(reservations) == 0:
        return f"{user_name}님의 {time_hint} 예약을 찾을 수 없습니다."

    if len(reservations) > 1:
        return f"{user_name}님의 {time_hint} 예약이 여러 개입니다. 더 구체적인 정보를 제공해주세요."

    reservation = reservations[0]
    reservation_id = reservation['reservation_id']

    # 수정할 필드만 포함하는 딕셔너리 생성
    update_kwargs = {}

    if start_datetime:
        try:
            update_kwargs['start_datetime'] = datetime.strptime(start_datetime, "%Y-%m-%d %H:%M")
        except ValueError:
            return f"start_datetime의 날짜 형식이 올바르지 않습니다. 'YYYY-MM-DD HH:MM' 형식을 사용해주세요."

    if end_datetime:
        try:
            update_kwargs['end_datetime'] = datetime.strptime(end_datetime, "%Y-%m-%d %H:%M")
        except ValueError:
            return f"end_datetime의 날짜 형식이 올바르지 않습니다. 'YYYY-MM-DD HH:MM' 형식을 사용해주세요."

    if title:
        update_kwargs['title'] = title

    if purpose:
        update_kwargs['purpose'] = purpose

    # 수정할 내용이 없으면 에러
    if not update_kwargs:
        return "수정할 내용이 없습니다. start_datetime, end_datetime, title, purpose 중 하나 이상을 제공해주세요."

    success, message = update_reservation(reservation_id, **update_kwargs)
    return message

@tool
def cancel_user_reservation(time_hint: str, user_name: str) -> str:
    """사용자의 예약을 취소합니다. 시간 힌트로 예약을 찾습니다.

    Args:
        time_hint: 시간 힌트 (예: "6시", "7시")
        user_name: 예약자 이름

    Returns:
        취소 결과 메시지
    """
    reservations = find_user_reservations(user_name, time_hint)

    if len(reservations) == 0:
        return f"{user_name}님의 {time_hint} 예약을 찾을 수 없습니다."

    if len(reservations) > 1:
        return f"{user_name}님의 {time_hint} 예약이 여러 개입니다. 더 구체적인 정보를 제공해주세요."

    reservation = reservations[0]
    reservation_id = reservation['reservation_id']

    success, message = cancel_reservation(reservation_id)
    return message

@tool
def list_user_reservations(user_name: str) -> str:
    """사용자의 모든 예약을 조회합니다. "내 예약", "나의 예약" 등의 표현이 나오면 이 tool을 사용하세요.

    Args:
        user_name: 예약자 이름 (필수). 사용자가 "내", "나의" 등의 표현을 사용하면 현재 세션의 user_name을 사용합니다.

    Returns:
        예약 목록을 문자열로 반환
    """
    reservations = find_user_reservations(user_name)

    if len(reservations) == 0:
        return f"{user_name}님의 예약이 없습니다."

    result = f"{user_name}님의 예약 목록:\n"
    for res in reservations:
        # 건물 이름 찾기
        building_name = None
        for name, bid in building_ids.items():
            if bid == res['building_id']:
                building_name = name
                break

        # 실제 층 번호 찾기 (내부 층 번호에서 실제 층 번호로)
        actual_floor = None
        if res['building_id'] in floor_ids:
            for floor_key, floor_data in floor_ids[res['building_id']].items():
                if floor_data[0] == res['floor_id']:  # floor_data[0]이 내부 층 번호
                    actual_floor = floor_key
                    break

        # 회의실 이름 찾기
        room_name = None
        if res['building_id'] in floor_ids and actual_floor is not None:
            rooms = floor_ids[res['building_id']][actual_floor][1]
            for rname, rid in rooms.items():
                if rid == res['room_id']:
                    room_name = rname
                    break

        # 회의실 정보 포맷팅
        room_info = ""
        if building_name and actual_floor is not None and room_name:
            room_info = f" ({building_name} {actual_floor}층 {room_name})"
        elif room_name:
            room_info = f" ({room_name})"

        result += f"- {res['title']}{room_info}: {res['start_datetime'].strftime('%Y-%m-%d %H:%M')} ~ {res['end_datetime'].strftime('%H:%M')}\n"

    return result

# Tools 리스트
tools = [list_rooms, create_reservation, update_user_reservation, cancel_user_reservation, list_user_reservations]

# LLM에 tools 바인딩
llm_with_tools = llm.bind_tools(tools)

# Agent 실행 함수
def run_agent(user_query: str, user_name: str = "사용자"):
    """회의실 Agent를 실행합니다."""
    messages = [
        SystemMessage(content=f"""당신은 회의실 예약을 도와주는 AI 어시스턴트입니다.

현재 세션의 사용자 이름: {user_name}

주요 기능:
1. 회의실 조회: 특정 건물의 특정 층에 있는 회의실 목록을 조회할 수 있습니다.
2. 회의실 예약: 회의실을 예약할 수 있습니다. 날짜와 시간을 명확히 파악해야 합니다.
3. 예약 조회: 사용자의 예약 목록을 조회할 수 있습니다. "내 예약", "나의 예약" 등의 표현이 나오면 list_user_reservations tool을 사용하세요.
4. 예약 수정: 사용자의 예약을 수정할 수 있습니다. 시간 힌트(예: "6시", "7시")로 예약을 찾습니다.
5. 예약 삭제: 사용자의 예약을 취소할 수 있습니다.

중요 규칙:
- 사용자가 "내", "나의", "내 예약" 등의 표현을 사용하면 반드시 현재 세션의 user_name({user_name})을 사용해야 합니다.
- list_user_reservations, update_user_reservation, cancel_user_reservation, create_reservation tool을 사용할 때는 항상 user_name 파라미터에 {user_name}을 전달해야 합니다.
- 건물 이름은 "sku-tower", "skt-tower" 등으로 입력됩니다.
- 층 번호는 실제 층 번호를 사용합니다 (예: 24층).
- 예약 시 날짜가 명시되지 않으면 오늘 날짜를 사용합니다.
- 시간 힌트(예: "6시 회의")로 예약을 찾을 때는 해당 시간에 시작하는 예약을 찾습니다.

항상 적절한 tool을 사용하여 사용자의 요청을 처리하세요. 사용자가 예약을 조회하거나 수정/삭제할 때는 반드시 user_name을 tool에 전달해야 합니다."""),
        HumanMessage(content=user_query)
    ]

    max_iterations = 10
    iteration = 0

    while iteration < max_iterations:
        iteration += 1

        # LLM 호출
        response = llm_with_tools.invoke(messages)
        messages.append(response)

        # Tool 호출이 없으면 종료
        if not response.tool_calls:
            return response.content

        # Tool 실행
        tool_map = {t.name: t for t in tools}

        for tool_call in response.tool_calls:
            func_name = tool_call["name"]
            args = tool_call["args"]

            # user_name이 필요한 함수에 user_name 추가 (없으면 무조건 추가)
            if func_name in ["create_reservation", "update_user_reservation", "cancel_user_reservation", "list_user_reservations"]:
                current_user_name = args.get("user_name", "")
                if "user_name" not in args or not current_user_name or current_user_name == "사용자" or (isinstance(current_user_name, str) and current_user_name.strip() == ""):
                    args["user_name"] = user_name

            # 날짜가 없으면 오늘 날짜 사용
            if func_name == "create_reservation":
                if "start_datetime" not in args or not args["start_datetime"]:
                    # 시간만 추출
                    time_match = re.search(r'(\d+)시', user_query)
                    if time_match:
                        hour = int(time_match.group(1))
                        today = datetime.now().date()
                        args["start_datetime"] = f"{today} {hour:02d}:00"
                        args["end_datetime"] = f"{today} {hour+1:02d}:00"

            # update_user_reservation에서 날짜/시간 파싱 도와주기
            if func_name == "update_user_reservation":
                # time_hint로 예약을 찾아서 기존 날짜 가져오기
                time_hint = args.get("time_hint", "")
                current_user = args.get("user_name", user_name)

                # 기존 예약 찾기
                existing_reservations = find_user_reservations(current_user, time_hint)
                if existing_reservations and len(existing_reservations) > 0:
                    existing_res = existing_reservations[0]
                    existing_date = existing_res['start_datetime'].date()
                    existing_start = existing_res['start_datetime']
                    existing_end = existing_res['end_datetime']
                    duration = existing_end - existing_start

                    # start_datetime이 없으면 시간만 파싱해서 기존 날짜와 조합
                    if "start_datetime" not in args or not args.get("start_datetime"):
                        # "오후 7시", "7시" 등의 표현 파싱
                        time_match = re.search(r'(\d+)시', user_query)
                        if time_match:
                            hour_12 = int(time_match.group(1))

                            # 오후/오전 키워드 확인
                            user_query_lower = user_query.lower()
                            is_pm = any(keyword in user_query_lower for keyword in ['오후', 'pm', '저녁', '밤', '야간'])
                            is_am = any(keyword in user_query_lower for keyword in ['오전', 'am', '아침', '새벽'])

                            # 24시간 형식으로 변환
                            if is_pm:
                                if hour_12 == 12:
                                    hour_24 = 12
                                else:
                                    hour_24 = hour_12 + 12
                            elif is_am:
                                if hour_12 == 12:
                                    hour_24 = 0
                                else:
                                    hour_24 = hour_12
                            else:
                                # 키워드가 없으면 오후 시간으로 가정 (일반적으로 회의는 오후에 많음)
                                if 1 <= hour_12 <= 11:
                                    hour_24 = hour_12 + 12
                                else:
                                    hour_24 = hour_12

                            # 기존 날짜와 새로운 시간 조합
                            new_start = datetime.combine(existing_date, datetime.min.time().replace(hour=hour_24, minute=0))
                            new_end = new_start + duration

                            args["start_datetime"] = new_start.strftime("%Y-%m-%d %H:%M")
                            args["end_datetime"] = new_end.strftime("%Y-%m-%d %H:%M")

            selected_tool = tool_map.get(func_name)
            if selected_tool is None:
                tool_msg = ToolMessage(
                    content=f"Unknown tool: {func_name}",
                    tool_call_id=tool_call.get("id")
                )
            else:
                try:
                    tool_result = selected_tool.invoke(args)
                    tool_msg = ToolMessage(
                        content=str(tool_result),
                        tool_call_id=tool_call.get("id")
                    )
                except Exception as e:
                    tool_msg = ToolMessage(
                        content=f"Error: {str(e)}",
                        tool_call_id=tool_call.get("id")
                    )

            messages.append(tool_msg)

    return "최대 반복 횟수에 도달했습니다."

# 테스트
print("회의실 Agent가 준비되었습니다!")
print("사용 예시:")
print("1. run_agent('sku타워 24층에 회의실이 뭐가 있지?')")
print("2. run_agent('sku타워 24층 회의실 예약해줘', user_name='홍길동')")
print("3. run_agent('내 6시 회의 수정해줘', user_name='홍길동')")
print("4. run_agent('내 7시 회의 삭제해줘', user_name='홍길동')")
```

# 평가 모듈 (Evaluation Module)

이 모듈은 회의실 Agent의 성능을 평가합니다.
- **Agent 비교 평가**: 두 agent의 응답을 비교하여 어느 것이 더 우수한지 판단
- **Plan 평가**: Agent가 실행한 계획(tool 호출 순서, 전략 등)에 대한 평가


```python
# ============================================
# 평가 모듈 (Evaluation Module)
# ============================================

from pydantic import BaseModel, Field
from typing import List, Optional, Dict, Any
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage

# 평가용 LLM (평가자 역할)
evaluator_llm = ChatOpenAI(
    model="gpt-4o",
    temperature=0.1,
    api_key=api_key,
)

# ============================================
# 1. Agent 비교 평가
# ============================================

class CriterionScore(BaseModel):
    """평가 기준별 점수"""
    criterion: str = Field(description="평가 기준 이름")
    score: float = Field(description="점수 (0.0 ~ 1.0)", ge=0.0, le=1.0)

class AgentComparisonResult(BaseModel):
    """Agent 비교 평가 결과"""
    winner: str = Field(description="더 우수한 agent (agent1 또는 agent2)")
    winner_score: float = Field(description="승자 점수 (0.0 ~ 1.0)", ge=0.0, le=1.0)
    loser_score: float = Field(description="패자 점수 (0.0 ~ 1.0)", ge=0.0, le=1.0)
    reasoning: str = Field(description="평가 근거 및 상세 설명")
    criteria_scores: List[CriterionScore] = Field(default_factory=list, description="세부 평가 항목별 점수 리스트")

def compare_agents(
    query: str,
    agent1_response: str,
    agent2_response: str,
    agent1_name: str = "Agent 1",
    agent2_name: str = "Agent 2",
    context: Optional[str] = None
) -> AgentComparisonResult:
    """두 agent의 응답을 비교하여 어느 것이 더 우수한지 평가합니다.

    Args:
        query: 사용자 쿼리
        agent1_response: 첫 번째 agent의 응답
        agent2_response: 두 번째 agent의 응답
        agent1_name: 첫 번째 agent 이름
        agent2_name: 두 번째 agent 이름
        context: 추가 컨텍스트 정보 (선택)

    Returns:
        AgentComparisonResult: 평가 결과
    """

    evaluation_prompt = f"""당신은 회의실 예약 Agent의 성능을 평가하는 전문가입니다.

사용자 쿼리: {query}

{agent1_name}의 응답:
{agent1_response}

{agent2_name}의 응답:
{agent2_response}

{f'추가 컨텍스트: {context}' if context else ''}

다음 기준으로 두 agent를 평가하세요:

1. **정확성 (Accuracy)**: 요청을 정확히 이해하고 올바른 정보를 제공했는가?
2. **완전성 (Completeness)**: 필요한 모든 정보를 포함했는가?
3. **사용자 친화성 (User-friendliness)**: 응답이 명확하고 이해하기 쉬운가?
4. **효율성 (Efficiency)**: 불필요한 정보 없이 핵심만 전달했는가?
5. **맥락 이해 (Context Understanding)**: 사용자의 의도를 정확히 파악했는가?

각 기준에 대해 0.0 ~ 1.0 점수를 부여하고, 전체적으로 더 우수한 agent를 선택하세요.
두 agent가 비슷한 수준이면 더 세밀한 기준으로 판단하세요.

반드시 criteria_scores 필드에 각 평가 기준별 점수를 리스트 형태로 제공하세요.
예: [{{"criterion": "정확성", "score": 0.9}}, {{"criterion": "완전성", "score": 0.8}}, ...]"""

    messages = [
        SystemMessage(content="당신은 회의실 예약 시스템 전문가이자 평가 전문가입니다. 객관적이고 공정하게 평가하세요."),
        HumanMessage(content=evaluation_prompt)
    ]

    # Structured output으로 평가 결과 받기
    model_with_structure = evaluator_llm.with_structured_output(AgentComparisonResult)
    result = model_with_structure.invoke(messages)

    return result

# ============================================
# 2. Plan 평가
# ============================================

class PlanEvaluationResult(BaseModel):
    """Plan 평가 결과"""
    overall_score: float = Field(description="전체 점수 (0.0 ~ 1.0)", ge=0.0, le=1.0)
    plan_quality: float = Field(description="계획의 질 (0.0 ~ 1.0)", ge=0.0, le=1.0)
    execution_efficiency: float = Field(description="실행 효율성 (0.0 ~ 1.0)", ge=0.0, le=1.0)
    tool_usage: float = Field(description="Tool 사용 적절성 (0.0 ~ 1.0)", ge=0.0, le=1.0)
    reasoning: str = Field(description="평가 근거 및 개선 제안")
    strengths: List[str] = Field(description="계획의 강점")
    weaknesses: List[str] = Field(description="계획의 약점")
    suggestions: List[str] = Field(description="개선 제안")

def evaluate_plan(
    query: str,
    tool_calls: List[Dict[str, Any]],
    final_response: str,
    execution_time: Optional[float] = None
) -> PlanEvaluationResult:
    """Agent가 실행한 계획(tool 호출 순서 및 전략)을 평가합니다.

    Args:
        query: 사용자 쿼리
        tool_calls: 실행된 tool 호출 목록 (각 항목은 tool_name, args, result 포함)
        final_response: 최종 응답
        execution_time: 실행 시간 (초, 선택)

    Returns:
        PlanEvaluationResult: 평가 결과
    """

    # Tool 호출 정보 포맷팅
    tool_calls_str = ""
    for i, tool_call in enumerate(tool_calls, 1):
        tool_name = tool_call.get('tool_name', 'Unknown')
        args = tool_call.get('args', {})
        result = tool_call.get('result', 'N/A')
        tool_calls_str += f"\n{i}. Tool: {tool_name}\n"
        tool_calls_str += f"   Args: {args}\n"
        tool_calls_str += f"   Result: {str(result)[:200]}...\n"

    evaluation_prompt = f"""당신은 회의실 예약 Agent의 실행 계획을 평가하는 전문가입니다.

사용자 쿼리: {query}

실행된 Tool 호출 순서:
{tool_calls_str}

최종 응답:
{final_response}

{f'실행 시간: {execution_time:.2f}초' if execution_time else ''}

다음 기준으로 계획을 평가하세요:

1. **계획의 질 (Plan Quality)**:
   - Tool 호출 순서가 논리적인가?
   - 필요한 정보를 단계적으로 수집했는가?
   - 불필요한 tool 호출이 없는가?

2. **실행 효율성 (Execution Efficiency)**:
   - 최소한의 tool 호출로 목표를 달성했는가?
   - 중복된 작업이 없는가?
   - 실행 시간이 적절한가?

3. **Tool 사용 적절성 (Tool Usage)**:
   - 각 tool을 올바르게 사용했는가?
   - 적절한 파라미터를 전달했는가?
   - Tool의 결과를 제대로 활용했는가?

각 기준에 대해 0.0 ~ 1.0 점수를 부여하고, 강점과 약점을 분석한 후 개선 제안을 제공하세요."""

    messages = [
        SystemMessage(content="당신은 Agent 실행 계획을 평가하는 전문가입니다. 객관적이고 건설적인 피드백을 제공하세요."),
        HumanMessage(content=evaluation_prompt)
    ]

    # Structured output으로 평가 결과 받기
    model_with_structure = evaluator_llm.with_structured_output(PlanEvaluationResult)
    result = model_with_structure.invoke(messages)

    return result

# ============================================
# 3. 통합 평가 함수
# ============================================

def evaluate_agent_performance(
    query: str,
    agent_response: str,
    tool_calls: List[Dict[str, Any]],
    expected_result: Optional[str] = None,
    execution_time: Optional[float] = None
) -> Dict[str, Any]:
    """Agent의 전체 성능을 평가합니다.

    Args:
        query: 사용자 쿼리
        agent_response: Agent의 최종 응답
        tool_calls: 실행된 tool 호출 목록
        expected_result: 기대 결과 (선택, 있으면 정확성 평가에 사용)
        execution_time: 실행 시간 (초, 선택)

    Returns:
        평가 결과 딕셔너리
    """
    results = {}

    # Plan 평가
    plan_eval = evaluate_plan(query, tool_calls, agent_response, execution_time)
    results['plan_evaluation'] = plan_eval

    # 기대 결과가 있으면 정확성 평가 추가
    if expected_result:
        accuracy_eval = compare_agents(
            query=query,
            agent1_response=agent_response,
            agent2_response=expected_result,
            agent1_name="실제 Agent",
            agent2_name="기대 결과",
            context="기대 결과와 실제 응답을 비교하여 정확성을 평가합니다."
        )
        results['accuracy_evaluation'] = accuracy_eval

    return results

print("평가 모듈이 준비되었습니다!")
print("사용 예시:")
print("1. compare_agents(query, agent1_response, agent2_response)")
print("2. evaluate_plan(query, tool_calls, final_response)")
print("3. evaluate_agent_performance(query, agent_response, tool_calls)")
```

```python
# ============================================
# 예제 3: 실제 Agent 실행 평가 (run_agent와 연동)
# ============================================

import time

def run_agent_with_tracking(user_query: str, user_name: str = "사용자"):
    """Tool 호출을 추적하는 버전의 run_agent

    Returns:
        tuple: (final_response, tool_calls_history, execution_time)
    """
    tool_calls_history = []
    start_time = time.time()

    messages = [
        SystemMessage(content=f"""당신은 회의실 예약을 도와주는 AI 어시스턴트입니다.

현재 세션의 사용자 이름: {user_name}

주요 기능:
1. 회의실 조회: 특정 건물의 특정 층에 있는 회의실 목록을 조회할 수 있습니다.
2. 회의실 예약: 회의실을 예약할 수 있습니다. 날짜와 시간을 명확히 파악해야 합니다.
3. 예약 조회: 사용자의 예약 목록을 조회할 수 있습니다. "내 예약", "나의 예약" 등의 표현이 나오면 list_user_reservations tool을 사용하세요.
4. 예약 수정: 사용자의 예약을 수정할 수 있습니다. 시간 힌트(예: "6시", "7시")로 예약을 찾습니다.
5. 예약 삭제: 사용자의 예약을 취소할 수 있습니다.

중요 규칙:
- 사용자가 "내", "나의", "내 예약" 등의 표현을 사용하면 반드시 현재 세션의 user_name({user_name})을 사용해야 합니다.
- list_user_reservations, update_user_reservation, cancel_user_reservation, create_reservation tool을 사용할 때는 항상 user_name 파라미터에 {user_name}을 전달해야 합니다.
- 건물 이름은 "sku-tower", "skt-tower" 등으로 입력됩니다.
- 층 번호는 실제 층 번호를 사용합니다 (예: 24층).
- 예약 시 날짜가 명시되지 않으면 오늘 날짜를 사용합니다.
- 시간 힌트(예: "6시 회의")로 예약을 찾을 때는 해당 시간에 시작하는 예약을 찾습니다.

항상 적절한 tool을 사용하여 사용자의 요청을 처리하세요. 사용자가 예약을 조회하거나 수정/삭제할 때는 반드시 user_name을 tool에 전달해야 합니다."""),
        HumanMessage(content=user_query)
    ]

    max_iterations = 10
    iteration = 0

    while iteration < max_iterations:
        iteration += 1

        # LLM 호출
        response = llm_with_tools.invoke(messages)
        messages.append(response)

        # Tool 호출이 없으면 종료
        if not response.tool_calls:
            execution_time = time.time() - start_time
            return response.content, tool_calls_history, execution_time

        # Tool 실행
        tool_map = {t.name: t for t in tools}

        for tool_call in response.tool_calls:
            func_name = tool_call["name"]
            args = tool_call["args"].copy()  # 원본 수정 방지

            # user_name이 필요한 함수에 user_name 추가
            if func_name in ["create_reservation", "update_user_reservation", "cancel_user_reservation", "list_user_reservations"]:
                current_user_name = args.get("user_name", "")
                if "user_name" not in args or not current_user_name or current_user_name == "사용자" or (isinstance(current_user_name, str) and current_user_name.strip() == ""):
                    args["user_name"] = user_name

            # 날짜가 없으면 오늘 날짜 사용
            if func_name == "create_reservation":
                if "start_datetime" not in args or not args["start_datetime"]:
                    time_match = re.search(r'(\d+)시', user_query)
                    if time_match:
                        hour = int(time_match.group(1))
                        today = datetime.now().date()
                        args["start_datetime"] = f"{today} {hour:02d}:00"
                        args["end_datetime"] = f"{today} {hour+1:02d}:00"

            # update_user_reservation에서 날짜/시간 파싱 도와주기
            if func_name == "update_user_reservation":
                time_hint = args.get("time_hint", "")
                current_user = args.get("user_name", user_name)

                existing_reservations = find_user_reservations(current_user, time_hint)
                if existing_reservations and len(existing_reservations) > 0:
                    existing_res = existing_reservations[0]
                    existing_date = existing_res['start_datetime'].date()
                    existing_start = existing_res['start_datetime']
                    existing_end = existing_res['end_datetime']
                    duration = existing_end - existing_start

                    if "start_datetime" not in args or not args.get("start_datetime"):
                        time_match = re.search(r'(\d+)시', user_query)
                        if time_match:
                            hour_12 = int(time_match.group(1))

                            user_query_lower = user_query.lower()
                            is_pm = any(keyword in user_query_lower for keyword in ['오후', 'pm', '저녁', '밤', '야간'])
                            is_am = any(keyword in user_query_lower for keyword in ['오전', 'am', '아침', '새벽'])

                            if is_pm:
                                if hour_12 == 12:
                                    hour_24 = 12
                                else:
                                    hour_24 = hour_12 + 12
                            elif is_am:
                                if hour_12 == 12:
                                    hour_24 = 0
                                else:
                                    hour_24 = hour_12
                            else:
                                if 1 <= hour_12 <= 11:
                                    hour_24 = hour_12 + 12
                                else:
                                    hour_24 = hour_12

                            new_start = datetime.combine(existing_date, datetime.min.time().replace(hour=hour_24, minute=0))
                            new_end = new_start + duration

                            args["start_datetime"] = new_start.strftime("%Y-%m-%d %H:%M")
                            args["end_datetime"] = new_end.strftime("%Y-%m-%d %H:%M")

            selected_tool = tool_map.get(func_name)
            if selected_tool is None:
                tool_result = f"Unknown tool: {func_name}"
                tool_msg = ToolMessage(
                    content=tool_result,
                    tool_call_id=tool_call.get("id")
                )
            else:
                try:
                    tool_result = selected_tool.invoke(args)
                    tool_msg = ToolMessage(
                        content=str(tool_result),
                        tool_call_id=tool_call.get("id")
                    )
                except Exception as e:
                    tool_result = f"Error: {str(e)}"
                    tool_msg = ToolMessage(
                        content=tool_result,
                        tool_call_id=tool_call.get("id")
                    )

            # Tool 호출 기록
            tool_calls_history.append({
                "tool_name": func_name,
                "args": args,
                "result": tool_result
            })

            messages.append(tool_msg)

    execution_time = time.time() - start_time
    return "최대 반복 횟수에 도달했습니다.", tool_calls_history, execution_time

print("run_agent_with_tracking 함수가 준비되었습니다!")
```
